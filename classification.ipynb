```python
from scipy.io import loadmat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#np.set_printoptions(threshold=1000)
```

```python
# load the data from the .mat file

data = loadmat('data/sp1s_aa.mat')
```

sp1s_aa.mat is a Matlab format (*.mat) containing variables:   
> clab: electrode labels  
> x_train: training trials (time x channels x trials)  
> y_train: corresponding labels (0: left, 1: right)  
> x_test: test trials (time x channels x trials)

```python
# check the dictionary keys to ensure correct variables loaded

data.keys()
```

```python
# check the shapes
print('Shape of clab: ', data['clab'].shape)
print('Shape of x_train: ', data['x_train'].shape)
print('Shape of y_train: ', data['y_train'].shape)
print('Shape of x_test: ', data['x_test'].shape)
```

```python
# create new variables from keys

clab = data['clab']
x_train = data['x_train']
y_train = data['y_train']
x_test = data['x_test']
```
